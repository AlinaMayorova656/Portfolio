#### 1.3. Анализ числа регистраций
Одна из важнейших метрик продукта — число зарегистрированных пользователей. Используя исторические данные, визуализируйте, как менялось число регистраций в приложении за время его существования.

- Агрегируйте исторические данные и рассчитайте число уникальных пользователей и число зарегистрированных пользователей для каждого дня наблюдения. Для простоты считайте, что у пользователя в течение дня бывает одна сессия максимум и статус регистрации в течение одного дня не может измениться.

- Постройте линейные графики общего числа пользователей и общего числа зарегистрированных пользователей по дням. Отобразите их на одном графике.

- Постройте отдельный линейный график доли зарегистрированных пользователей от всех пользователей по дням.

- На обоих графиках должны быть заголовок, подписанные оси X и Y, сетка и легенда.

````python
# Считаем общее количество уникальных пользователей за день
daily_users = sessions_history.groupby('session_date')['user_id'].nunique().reset_index()
daily_users.columns = ['session_date', 'total_users']
````

````python
# Считаем количество зарегистрированных пользователей за день
registered_sessions = sessions_history[sessions_history['registration_flag'] == 1]
daily_registered = registered_sessions.groupby('session_date')['user_id'].nunique().reset_index()
daily_registered.columns = ['session_date', 'registered_users']
````

````python
# Объединяем две таблицы
daily_stats = pd.merge(daily_users, daily_registered, on='session_date', how='left')
daily_stats
````

````python
daily_stats['session_date'] = pd.to_datetime(daily_stats['session_date'])

# Создаем график
fig, ax = plt.subplots(figsize=(15, 6))

# Строим линии
ax.plot(daily_stats['session_date'], 
        daily_stats['total_users'], 
        label='Все пользователи', 
        linewidth=2, 
        color='blue')
ax.plot(daily_stats['session_date'], 
        daily_stats['registered_users'], 
        label='Зарегистрированные пользователи',
        linewidth=2, 
        color='lightblue')

ax.set_title('Динамика количества пользователей по дням', fontsize=14, pad=20)
ax.set_xlabel('Дата', fontsize=12)
ax.set_ylabel('Количество пользователей', fontsize=12)
ax.legend(fontsize=12)
ax.grid(True, alpha=0.3)

ax.xaxis.set_major_formatter(mdates.DateFormatter('%d.%m.%Y'))
fig.autofmt_xdate(rotation=45)

plt.tight_layout()
plt.show()
````

**Наблюдения:**
- Общее количество пользователей снижается с 10.09.2025 - это критически важный тренд
- Зарегистрированные пользователи остаются стабильными в этот период снижения
- Получается, мы теряем преимущественно незарегистрированных пользователей

**Выводы:**
1. Приложение испытывает проблемы с удержанием незарегистрированных пользователей
2. Регистрация становится ключевым фактором удержания - зарегистрированные пользователи более лояльны
3. Необходимо срочно исследовать причины оттока незарегистрированных пользователей после 10.09.2025

````python
# Рассчитываем долю зарегистрированных пользователей
daily_stats['registration_rate'] = daily_stats['registered_users'] / daily_stats['total_users']

# Создаем график
fig, ax = plt.subplots(figsize=(12, 6))

# Строим график доли зарегистрированных пользователей
ax.plot(daily_stats['session_date'], daily_stats['registration_rate'],
        linewidth=2, color='darkblue')

ax.set_title('Доля зарегистрированных пользователей по дням', fontsize=14, pad=20)
ax.set_xlabel('Дата', fontsize=12)
ax.set_ylabel('Доля зарегистрированных пользователей', fontsize=12)
ax.grid(True, alpha=0.3)

# Настраиваем отображение дат на оси X
ax.xaxis.set_major_formatter(mdates.DateFormatter('%d.%m.%Y'))
fig.autofmt_xdate(rotation=45)

# Форматируем ось Y в процентах
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f'{y:.1%}'))

plt.tight_layout()
plt.show()
````

**Наблюдения:**
- Доля регистраций растет, особенно заметно в период снижения общего числа пользователей
**Это объяснимо:** если общее число пользователей снижается, а количество зарегистрированных стабильно, доля автоматически увеличивается

**Вывод:**
1. Рост доли регистраций - следствие оттока незарегистрированных пользователей, а не успешной работы над конверсией

#### 2.1 Формулировка нулевой и альтернативной гипотез

Перед тем как проводить А/B-тест, необходимо сформулировать нулевую и альтернативную гипотезы. Напомним изначальное условие: команда разработчиков рекомендательных систем создала новый алгоритм, который, по их мнению, будет показывать более интересный контент для каждого пользователя.

О какой метрике идёт речь? Как она будет учтена в формулировке гипотез?

**Нулевая гипотеза (H0):** Новый алгоритм рекомендаций не влияет на вовлеченность пользователей. Доля успешных сессий (с просмотром 4+ страниц) в тестовой группе (с новым алгоритмом) равна доле успешных сессий в контрольной группе (со старым алгоритмом).

*H0: p_control = p_test*

**Альтернативная гипотеза (H1):** Новый алгоритм рекомендаций увеличивает вовлеченность пользователей. Доля успешных сессий (с просмотром 4+ страниц) в тестовой группе (с новым алгоритмом) больше доли успешных сессий в контрольной группе (со старым алгоритмом).

*H1: p_control < p_test* 

Доля успешных сессий `good_session_rate` выбрана в качестве целевой метрики, потому что:
- Она непосредственно отражает удовлетворенность пользователей контентом
- Это прокси-метрика для вовлеченности, которая в перспективе влияет на регистрацию и монетизацию
- Из анализа исторических данных видно, что текущий уровень стабилен (~30%), что делает его хорошим базисом для сравнения
- Проблема оттока незарегистрированных пользователей может быть решена через повышение вовлеченности

**Дополнительные метрики для мониторинга:**
1. Абсолютное количество незарегистрированных пользователей (чтобы отследить влияние на отток)
2. Конверсия в регистрацию (как вторичная метрика)
3. Среднее количество просмотренных страниц (для дополнительной валидации улучшения рекомендаций)

#### 2.2. Расчёт размера выборки
В рамках курса вы уже рассчитывали размеры выборки и  использовали для этого онлайн-калькулятор. В этом задании предлагаем воспользоваться готовым кодом и рассчитать необходимое для вашего эксперимента количество пользователей.

Для этого установите в коде ниже следующие параметры:

- Уровень значимости — 0.05.

- Вероятность ошибки второго рода — 0.2.

- Мощность теста.

- Минимальный детектируемый эффект, или MDE, — 3%. Обратите внимание, что здесь нужно указать десятичную дробь, а не процент.

При расчёте размера выборки используйте метод `solve_power()` из класса `power.NormalIndPower` модуля `statsmodels.stats`.

Запустите ячейку и изучите полученное значение.

````python
# Задаём параметры
alpha = 0.05  # Уровень значимости
beta = 0.2  # Ошибка второго рода, часто 1 - мощность
power = 0.8  # Мощность теста
p = 0.3 # Базовый уровень доли
mde = 0.03 * p # Минимальный детектируемый эффект
effect_size = proportion_effectsize(p, p + mde)

# Инициализируем класс NormalIndPower
power_analysis = NormalIndPower()

# Рассчитываем размер выборки
sample_size = power_analysis.solve_power(
    effect_size = effect_size,
    power = power,
    alpha = alpha,
    ratio = 1 # Равномерное распределение выборок
)

print(f"Необходимый размер выборки для каждой группы: {int(sample_size)}")
````

Для расчёта процентной разницы воспользуйтесь формулой:
$$P = 100 \cdot  \frac{|A − B|}{A}$$

````python
# Выгружаем файл sessions_project_test_part.csv
sessions_test_part = pd.read_csv('https://code.s3.yandex.net/datasets/sessions_project_test_part.csv')
````

В шаге 2.1 нулевая и альтернативная гипотезы сформулированы не для текущего шага, а для шага 4.4, где оценивается статистическая значимость изменения ключевой метрики.

В нулевой гипотезе обычно утверждают отсутствие эффекта или различий (например, "равны" или "не отличаются"), а альтернативная гипотеза (H₁) предполагают наличие эффекта или различия (например, "отличаются", "больше" или "меньше").
    
Если направление не задано, то:   
* H₀ - равны;   
* H₁ - отличаются.   

**Значит**:

- H₀ (нулевая гипотеза) — количество уникальных дневных сессий `не различается` между тестовой и контрольной группами.
- H₁ (альтернативная гипотеза) — количество уникальных дневных сессий `различается` между тестовой и контрольной группами.

    
На будущее: Значения nan, 0 и 1 первый признак некорректного расчета, это может указывать на проблемы в расчётах или наличие идентичных или вырожденных данных.

Перед проверкой статистического теста сформулируем гипотезы:

- **Нулевая гипотеза (H₀)**: Среднее количество сессий в день в группе A равно среднему в группе B  
  (группы сопоставимы по трафику)
  
- **Альтернативная гипотеза (H₁)**: Среднее количество сессий в день в группе A не равно среднему в группе B  
  (группы различаются по трафику)

Цель проверки: убедиться, что группы A и B сопоставимы по количеству сессий, что является важным условием корректности A/B-теста.

````python
# 1. Создаем датафрейм с количеством уникальных сессий по дням и группам
daily_sessions = sessions_test.groupby(['session_date', 'test_group'])['session_id'].nunique().reset_index()

# Группа A
group_a_sessions = daily_sessions[daily_sessions['test_group'] == 'A']['session_id']

# Группа B
group_b_sessions = daily_sessions[daily_sessions['test_group'] == 'B']['session_id']

alpha = 0.05
stat_ttest, p_value_ttest = ttest_ind(
    group_a_sessions,
    group_b_sessions,
    equal_var=False  # тест Уэлча 
)

print(f"p-value теста Уэлча = {p_value_ttest:.4f}")

if p_value_ttest > alpha:
    print("p-value > alpha")
    print("Не отвергаем нулевую гипотезу")
    print("Статистически значимых различий в количестве уникальных дневных сессий между группами A и В нет")
else:
    print("p-value ≤ alpha")
    print("Отвергаем нулевую гипотезу")
    print("Обнаружены статистически значимые различия в количестве уникальных дневных сессий между группами")
````

#### 4.4. Насколько статистически значимо изменение ключевой метрики

На предыдущем шаге вы убедились, что количество успешных сессий в тестовой выборке примерно на 1.1% выше, чем в контрольной, но делать выводы только на основе этого значения будет некорректно. Для принятия решения всегда необходимо отвечать на вопрос: является ли это изменение статистически значимым.

- Используя статистический тест, рассчитайте, является ли изменение в метрике доли успешных сессий статистически значимым.

- Выведите на экран полученное значение p-value и свои выводы о статистической значимости. Напомним, что уровень значимости в эксперименте был выбран на уровне 0.05.

````python
# Группа A (контрольная)
success_a = sessions_test[sessions_test['test_group'] == 'A']['good_session'].sum()
total_a = sessions_test[sessions_test['test_group'] == 'A'].shape[0]

# Группа B (тестовая)
success_b = sessions_test[sessions_test['test_group'] == 'B']['good_session'].sum()
total_b = sessions_test[sessions_test['test_group'] == 'B'].shape[0]

alpha = 0.05 

# z-тест для двух пропорций
stat_ttest, p_value_ttest = proportions_ztest(
    [success_a, success_b], 
    [total_a, total_b],    
    alternative='smaller'    # проверяем гипотезу p_A < p_B
)

print(f"p-value = {p_value_ttest:.6f}")

if p_value_ttest > alpha:
    print(f"p-value={p_value_ttest:.6f} > {alpha}")
    print("Нулевая гипотеза находит подтверждение!")
    print("Нет статистически значимых доказательств, что новый алгоритм лучше.")
else:
    print(f"p-value={p_value_ttest:.6f} < {alpha}")
    print("Нулевая гипотеза не находит подтверждения!")
    print("Новый алгоритм действительно увеличил долю успешных сессий.")
````
