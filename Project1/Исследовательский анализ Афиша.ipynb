3. Наличие аномальных значений: заказ с отрицательной выручкой в RUB (вероятно, возвраты билетов)
4. Экстремальные положительные выбросы в обеих валютах

**Принятое решение: Удаление выбросов по 99-му процентилю** для каждой валюты отдельно


```python
# Сохраняем исходный размер
initial_size = len(df)

# Фильтруем данные
filtered_df = df[
    ((df['currency_code'] == 'rub') & (df['revenue'] <= outliers_rub)) |
    ((df['currency_code'] == 'kzt') & (df['revenue'] <= outliers_kzt))
]

print(f"Изначальное количество записей: {initial_size}")
print(f"Количество записей после фильтрации: {len(filtered_df)}")
print(f"Удалено выбросов: {initial_size - len(filtered_df)}")
```

Перед обработкой изучим отрицательные значения.

```python
# Получаем фильтр отрицательных значений
negative_mask = pd.to_numeric(filtered_df['revenue'], errors='coerce') < 0
# Используем  для фильтрации исходного датафрейма
negative_records = filtered_df[negative_mask]
print(f"Найдено отрицательных значений: {len(negative_records)}")
```

### Создание новых столбцов

Для корректного анализа финансовых показателей необходимо привести все заказы к единой валюте — российскому рублю, так как исходные данные содержат заказы в двух валютах: рубли (RUB) и казахстанские тенге (KZT). Используя датасет с ежедневными курсами тенге к рублю за 2024 год, где значение `curs` указано за 100 тенге, рассчитаем курс для 1 тенге и создадим новый столбец `revenue_rub` с выручкой в рублях для каждого заказа.

```python
# Приводим к типу дата и считаем курс для 1 тенге
currency['data'] = pd.to_datetime(currency['data'])
currency['rate'] = currency['curs'] / 100
```

```python
# Объединяем с курсами по дате
df = df.merge(currency[['data', 'rate']], left_on='created_dt_msk', right_on='data', how='left')
# Для заказов в RUB ставим курс 1
df['rate'] = df['rate'].fillna(1)
```

```python
# Создаём столбец revenue_rub
df['revenue_rub'] = df['revenue']  # копируем выручку
# Для заказов в KZT пересчитываем
kzt_mask = df['currency_code'] == 'kzt'
df.loc[kzt_mask, 'revenue_rub'] = df.loc[kzt_mask, 'revenue'] * df.loc[kzt_mask, 'rate']

# Удаляем вспомогательные столбцы
df = df.drop(columns=['data', 'rate'])
```

Для более детального анализа ценовой политики мероприятий и поведения пользователей рассчитаем выручку в расчете на один проданный билет. Этот показатель позволит сравнивать стоимость билетов на разные мероприятия независимо от количества билетов в заказе, а также выявлять изменения в ценовых сегментах. Создадим столбец `one_ticket_revenue_rub`, разделив общую выручку заказа в рублях на количество купленных билетов.

```python
# Создание столбца one_ticket_revenue_rub 
df['one_ticket_revenue_rub'] = df['revenue_rub'] / df['tickets_count']
```

Для анализа сезонности продаж и выявления трендов в пользовательской активности необходимо выделить месяц оформления заказа в отдельный столбец. Это позволит группировать данные по месяцам, отслеживать динамику выручки, количества заказов и других метрик в разрезе месяцев, а также сравнивать показатели между разными периодами. Создадим столбец `month`, извлекая номер месяца из даты создания заказа.

```python
# Создание столбца c месяцем оформления заказа
df['month'] = df['created_dt_msk'].dt.month
```

Для анализа влияния сезонности на продажи и выявления изменений в пользовательских предпочтениях между летним и осенним периодами 2024 года создадим столбец `season`.

```python
# Функция определяющая сезон
def get_season(month):
    if month in [12, 1, 2]:
        return 'зима'
    elif month in [3, 4, 5]:
        return 'весна'
    elif month in [6, 7, 8]:
        return 'лето'
    else:  # month in [9, 10, 11]
        return 'осень'
```

```python
# Создаём столбец и преобразуем тип данных
df['season'] = df['month'].apply(get_season)
df['season'] = df['season'].astype('category')
```

```python
# Рассчитываем среднюю стоимость билета 
summer_avg = df_seasons[df_seasons['season'] == 'лето'].groupby('event_type_main')['one_ticket_revenue_rub'].mean()
autumn_avg = df_seasons[df_seasons['season'] == 'осень'].groupby('event_type_main')['one_ticket_revenue_rub'].mean()
```

```python
# Создаём датафрейм DataFrame
price_comparison = pd.DataFrame({
    'Лето': summer_avg,
    'Осень': autumn_avg,
    'Изменение, %': ((autumn_avg - summer_avg) / summer_avg * 100)
}).fillna(0)
pd.set_option('display.float_format', '{:.2f}'.format)

# Отсортируем датафрейм
price_comparison = price_comparison.sort_values('Изменение, %', ascending=False)
price_comparison
```

```python
# Визуализация с средней стоимостью билета по типу мероприятия
price_comparison[['Лето', 'Осень']].plot(kind='bar',
                                          color=['orange', 'brown'],
                                          edgecolor='black',
                                          alpha=0.8,
                                          figsize=(18, 10))

plt.title('Средняя стоимость билета по типам мероприятий: лето vs осень', fontsize=18)
plt.xlabel('Тип мероприятия', fontsize=16)
plt.ylabel('Средняя стоимость билета, руб.', fontsize=16)
plt.legend(title='Сезон', fontsize=11, title_fontsize=12)
plt.xticks(rotation=0, fontsize=12)
plt.grid(axis='y', alpha=0.3)

for p in plt.gca().patches:
    plt.text(p.get_x() + p.get_width()/2., p.get_height() + 2, 
             f'{p.get_height():.2f}', ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

Можно увидеть, что категории с подорожанием билетов `выставки`, `стендап` осенью демонстрируют низкую долю заказов `1.4%` и `4.1%` соответственно, в то время как подешевевшие театры и концерты остаются лидерами по популярности. Это может указывать на то, что **снижение цены является эффективным инструментом привлечения аудитории,** тогда как рост стоимости, даже незначительный, сдерживает спрос.

Для наглядной оценки того, насколько сильно изменилась стоимость билетов на разные типы мероприятий, построим столбчатую диаграмму относительного изменения в процентах. Это позволит сразу увидеть, какие категории подорожали, а какие подешевели, и сравнить масштаб изменений.

```python
# График: относительное изменение цены
plt.figure(figsize=(18, 10))

colors = ['green' if x > 0 else 'red' for x in price_comparison['Изменение, %']]
price_comparison['Изменение, %'].plot(kind='bar',
                                      color=colors,
                                      edgecolor='black',
                                      alpha=0.8)

plt.title('Относительное изменение средней стоимости билета от лета к осени', fontsize=18)
plt.xlabel('Тип мероприятия', fontsize=16)
plt.ylabel('Изменение, %', fontsize=16)
plt.axhline(y=0, color='black', linewidth=0.8, linestyle='--')
plt.xticks(rotation=0, fontsize=12)
plt.grid(axis='y', alpha=0.3)

# Добавляем значения
for i, v in enumerate(price_comparison['Изменение, %']):
    plt.text(i, v, f'{v:.1f}%', ha='center', va='bottom' if v >= 0 else 'top', fontsize=12)
    
plt.tight_layout()
plt.show()
```

Представленная диаграмма наглядно демонстрирует четкое разделение категорий на две группы:
1. Зеленая зона (рост цен):

2. Красная зона (снижение цен):
- Театр (-17.8%)
- Ёлки (-15.4%)
- Концерты (-12.0%) 
- Спорт и Другое — незначительное снижение (-1.6% и -1.7%)

**Наблюдения подтверждаются:** именно подешевевшие категории (театр, спорт, ёлки, концерты) остаются лидерами спроса, а подорожавшие (выставки, стендап) занимают минимальную долю заказов.

**Итоговый вывод:** С наступлением осени 2024 года наблюдается перераспределение спроса в пользу театра, спортивных событий и семейных мероприятий (0+), при этом появляется новая категория «Ёлки». Концерты теряют долю, но остаются самым популярным типом.

Средняя стоимость билета снизилась на 12–18% для лидирующих категорий (театр, концерты, ёлки), что совпало с ростом их популярности. Категории с повышением цены (выставки, стендап) занимают минимальную долю заказов.

Таким образом, подтверждается обратная зависимость между ценой и спросом: **снижение стоимости билетов на ключевые мероприятия способствует росту их популярности, а удорожание — сдерживает спрос.**

t-критерий Стьюдента в классическом виде (с предположением о равенстве дисперсий) неприменим. При больших размерах выборок (>1000) можно использовать **тест Уэлча** (не предполагает равенства дисперсий), который достаточно устойчив к отклонениям от нормальности. Для проверки робастности дополнительно применим непараметрический **тест Манна–Уитни.**

```python
# Выделяем выборки
desktop = user_orders_clean.loc[user_orders_clean['device_type_canonical'] == 'desktop', 'orders_count']
mobile = user_orders_clean.loc[user_orders_clean['device_type_canonical'] == 'mobile', 'orders_count']
```

```python
# Считаем размер эффекта
pooled_std = np.sqrt((desktop.std()**2 + mobile.std()**2) / 2)
effect_size = (mobile.mean() - desktop.mean()) / pooled_std
effect_size
```

```python
# Расчёт мощности теста 
analysis = TTestIndPower()
power = analysis.solve_power(
    effect_size=effect_size,
    nobs1=len(desktop),
    ratio=1,
    alpha=0.05,
    alternative='two-sided'
)
print(f"Мощность теста : {power:.2f}")
```

```python
# Тест Уэлча 
alpha = 0.05

stat_welch, p_value_welch = ttest_ind(mobile, desktop, equal_var=False, alternative='greater')

if p_value_welch < alpha:
    print(f"p-value = {p_value_welch:.3f} < alpha")
    print("Нулевая гипотеза отвергается.")
    print("Среднее число заказов на mobile пользователя статистически значимо выше, чем на desktop")
else:
    print(f"p-value = {p_value_welch:.3f} > alpha")
    print("Нулевая гипотеза не отвергается.")
    print("Нет статистически значимых доказательств, что mobile активнее.")
```

<div class="alert alert-warning">
    <h2> Комментарий ревьюера v1 <a class="tocSkip"> </h2>
    
<b>Небольшие замечания и рекомендации💡:</b>p-value - верояность, как следствие по своей сути не может быть = 0. В таких случаях правильнее применять экспоненциальную форму вывода результата

**Для непараметрического теста**, который сравнивает не средние, а целые распределения, **переформулируем гипотезы**:
1. Нулевая гипотеза H₀: распределения количества заказов в группах `mobile` и `desktop` статистически не различаются (нет систематического сдвига)
2. Альтернативная гипотеза H₁: распределение количества заказов в группе `mobile` имеет статистически значимый сдвиг в сторону больших значений относительно группы `desktop`

```python
# Тест Манна-Уитни 
alpha = 0.05

stat_mw, p_value_mw = mannwhitneyu(mobile, desktop, alternative='greater')

if p_value_mw < alpha:
    print(f"p-value = {p_value_mw:.3f} < alpha")
    print("Отвергаем H₀: распределение заказов в группе `mobile` значимо сдвинуто в сторону больших значений.")
else:
    print(f"p-value = {p_value_mw:.3f} > alpha")
    print("Не отвергаем H₀: распределения не различаются.")
```

**Итоговый вывод:**

Проведенные статистические тесты — параметрический **тест Уэлча** и непараметрический **тест Манна–Уитни** — дали согласованные результаты: 
- оба показали чрезвычайно низкие значения `p-value (< 0.001)`, что позволяет **уверенно отвергнуть нулевую гипотезу** об отсутствии различий между группами.

Это свидетельствует о том, что среднее количество заказов на мобильного пользователя статистически значимо выше, чем на десктопного, а распределение заказов в группе мобильных пользователей систематически сдвинуто в сторону больших значений.

Мощность теста, равная 1.00, подтверждает, что при таком объёме выборок вероятность ошибки II рода практически отсутствует.

**Таким образом, результаты обладают высокой статистической надёжностью и могут служить основанием для принятия продуктовых решений в пользу развития мобильного направления сервиса Яндекс Афиша.**

- Показывает значение, которое делит выборку пополам ("типичного" пользователя)
- Позволяет корректно сравнивать центральную тенденцию в группах с разным характером распределения

Тем не менее, для полноты анализа мы сначала проверим исходную гипотезу с помощью **теста Уэлча**, а затем проведём более корректный непараметрический тест.

<div class="alert alert-success">
<h2> Комментарий ревьюера v1 <a class="tocSkip"> </h2>

<b>Все супер!👍:</b>Предпосылки описаны, молодец, чт не забываешь про это

```python
# Формируем выборки
mobile_intervals = intervals_clean.loc[intervals_clean['device_type_canonical'] == 'mobile','days_since_prev']
desktop_intervals = intervals_clean.loc[intervals_clean['device_type_canonical'] == 'desktop','days_since_prev']
```

```python
# Рассчитываем размер эффекта
pooled_std = np.sqrt((desktop_intervals.std()**2 + mobile_intervals.std()**2) / 2)
effect_size = (mobile_intervals.mean() - desktop_intervals.mean()) / pooled_std
print(f"Размер эффекта : {effect_size:.3f}")
```

```python
# Расчет мощности теста
analysis = TTestIndPower()
power = analysis.solve_power(
    effect_size=effect_size,
    nobs1=len(desktop_intervals),
    ratio=len(mobile_intervals)/len(desktop_intervals), 
    alpha=0.05,
    alternative='two-sided'
)
print(f"Мощность теста: {power:.2f}")
```

Размер эффекта отрицательный (-0.146) и очень мал, что говорит о слабой практической значимости различий. Мощность теста близка к 100%, поэтому если бы эффект существовал в проверяемом направлении, мы бы его гарантированно обнаружили.

```python
# Тест Уэлча для сравнения средних
alpha = 0.05

stat, p_value = ttest_ind(mobile_intervals, desktop_intervals, 
                          equal_var=False, alternative='greater')

if p_value < 0.05:
    print(f"\np-value теста Уэлча: {p_value:.10f} < alpha")
    print("Нулевая гипотеза отвергается.")
    print("Среднее время между заказами у mobile статистически значимо ВЫШЕ, чем у desktop.")
else:
    print(f"\np-value теста Уэлча: {p_value:.10f} > alpha")
    print("Нулевая гипотеза не отвергается.")
    print("Нет статистически значимых доказательств, что среднее время у mobile выше.")
    print("Фактически, наблюдается обратная тенденция (mobile < desktop).")
```

**Вывод:** **Статистический тест не выявил значимых доказательств того, что среднее время между заказами у мобильных пользователей выше,** чем у десктопных (p-value ≈ 1.00). Наблюдаемые средние значения (13,8 дня у mobile против 18,1 дня у desktop) указывают на обратную тенденцию, **однако из-за сильной асимметрии распределений и большого количества нулевых интервалов в десктопной группе, сравнение средних не вполне корректно.**

Учитывая выявленные особенности данных (ненормальное распределение, наличие выбросов, большое количество нулей), **наиболее корректным является непараметрический тест Манна–Уитни.** Этот тест не требует нормальности распределения, устойчив к выбросам и сравнивает не средние значения, а целые распределения.

Переформулированные гипотезы для непараметрического теста:
1. Нулевая гипотеза H₀: распределения интервалов между заказами в группах` mobile` и `desktop` статистически не различаются (нет систематического сдвига)
2. Альтернативная гипотеза H₁: распределение интервалов в группе `mobile` имеет статистически значимый сдвиг относительно группы `desktop` (двусторонняя альтернатива)

```python
alpha = 0.05

stat, p_value_two = mannwhitneyu(mobile_intervals, desktop_intervals, alternative='two-sided')
print(f"Двусторонний тест")
print(f"p-value = {p_value_two:.10f}")

if p_value_two < alpha:
    print("Статистически значимые различия обнаружены. Нулевая гипотеза отвергается")
    print("Распределения интервалов в группах различаются.")
    
    if mobile_intervals.median() < desktop_intervals.median():
        print("Медиана интервалов у mobile меньше, чем у desktop.")
    else:
        print("Медиана интервалов у mobile больше, чем у desktop.")
else:
    print("Статистически значимых различий не обнаружено.")
    print("Распределения интервалов в группах не различаются.")
```

**Вывод:** Гипотеза о том, что интервалы между заказами у мобильных пользователей больше, чем у десктопных, **подтверждается** при корректном сравнении медиан с помощью непараметрического теста. **Мобильные пользователи не только активнее (совершают больше заказов), но и демонстрируют более регулярное возвращение на сервис** с типичным интервалом в 1 день, тогда как десктопные пользователи либо покупают несколько билетов за один раз, либо не проявляют повторной активности.

**Итоговый вывод:** Гипотеза о том, что среднее время между заказами у мобильных пользователей выше, чем у десктопных, не подтверждается. Однако более глубокий анализ с использованием корректных статистических методов показывает, что распределения интервалов значимо различаются, и мобильные пользователи демонстрируют более регулярное возвращение на сервис (медиана 1 день против 0 дней у desktop). Это полностью согласуется с первой гипотезой о том, что **мобильные пользователи активнее: они не только совершают больше заказов в целом, но и делают это регулярно, возвращаясь на сервис практически каждый день.**
